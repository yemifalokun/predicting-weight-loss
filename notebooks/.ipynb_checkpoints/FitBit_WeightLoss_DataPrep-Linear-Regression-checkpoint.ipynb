```python
# Read daily activity log info data file into DataFrame
dailyActivity_data_df = pd.read_csv("../data/Fitbitdata/dailyActivity_merged.csv")
dailyActivity_data_df.head(10)
```

```python
# Number of records in dataframe
dailyActivity_data_df.shape
```

```python
# Process this data file to calculate weight loss flag based on previous day calorie record
# https://www.appsloveworld.com/pandas/100/189/how-to-create-a-new-column-based-on-row-value-in-previous-row-in-pandas-dataframe

# Printing DataFrame after applying diff function
#print("Difference: ")
#print(dailyActivity_data_df.diff())

dailyActivity_data_df['weightloss'] = np.where(dailyActivity_data_df['Calories'].shift(1) > dailyActivity_data_df['Calories'],1,0)
dailyActivity_data_df
```

```python
# saving the dataframe
dailyActivity_data_df.to_csv('../data/Fitbitdata/dailyActivity_merged_processed.csv')
```

```python
# find out unique id count of daily activity records
dailyActivity_data_df['Id'].value_counts()
```

```python
# Print percentage of Columns that are null in relation to the total number of columns
for column in dailyActivity_data_df.columns:
    if dailyActivity_data_df[column].isnull().sum() > 0:
        print(column, ': {:.2%}'.format(dailyActivity_data_df[column].isnull().sum() /
                                               dailyActivity_data_df[column].shape[0]))
```
